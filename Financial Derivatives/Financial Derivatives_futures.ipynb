```python
def futures():
    import pandas as pd
    import math as m
    So = float(input("Value of the Stock at time 0 (So): "))
    r = float(input("intrest rate of a risk free asset: "))
    t = float(input("time of maturity (in years): "))
    k = float(input(f"Strike price of the future for the stock in {t} years: "))
    print()

    if k>So*(m.exp(r*t)):
        print("The future is Overpriced")
        print()
        print("-"*40)
        print("at t = 0")
        print()
        print(f"Borrow from bank:   {So}")
        print(f"Buy Stock:         ({So})")
        print(f"short forward contract for price {k}")
        print(f"Net Cashflow:         0")
        print()
        print("-"*40)
        print()
        print("-"*40)
        print(f"at t = {t}")
        print()
        print(f"Repay the Bank:     ({So*(m.exp(r*t)):.4f})")
        print(f"sell the stock at:   {k}")
        print("deliver on the short forward")
        print(f"Arbitrage profit:    {k-(So*(m.exp(r*t))):.4f}")
        print("-"*40)
        

    elif k<So*(m.exp(r*t)):
        print("The future is Underpriced")
        print()
        print("-"*40)
        print("at t = 0")
        print()
        print(f"lend to  bank:       ({So})")
        print(f"Short Stock:          {So}")
        print(f"long forward contract for price {k}")
        print(f"Net Cashflow:         0")
        print()
        print("-"*40)
        print()
        print("-"*40)
        print(f"at t = {t}")
        print()
        print(f"Recieve from Bank:      {So*(m.exp(r*t)):.4f}")
        print(f"buy the stock at:      ({k})")
        print("close the short position on short")
        print(f"Arbitrage profit:       {(So*(m.exp(r*t))-k):.4f}")
        print("-"*40)

    elif k == So*(m.exp(r*t)):
        print("The future is rightly priced, there is no arbitrage opportuinty")
```

```python
futures()
```
