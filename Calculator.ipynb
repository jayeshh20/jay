```python
import tkinter as tk

# Basic math functions
def add(x, y): return x + y
def sub(x, y): return x - y
def mul(x, y): return x * y
def div(x, y): return x / y if y != 0 else "Error"

# Function to handle button press
def press(num):
    global expression
    expression += str(num)
    equation.set(expression)

# Function to calculate result
def equalpress():
    global expression
    try:
        # Split expression into parts (number, operator, number)
        tokens = expression.split()
        if len(tokens) == 3:
            x, op, y = tokens
            x, y = float(x), float(y)
            if op == "+": result = add(x, y)
            elif op == "-": result = sub(x, y)
            elif op == "*": result = mul(x, y)
            elif op == "/": result = div(x, y)
            else: result = "Error"
        else:
            result = "Error"
        equation.set(str(result))
        expression = str(result)
    except:
        equation.set("Error")
        expression = ""

# Function to clear
def clear():
    global expression
    expression = ""
    equation.set("")

# --- Main program ---
expression = ""
root = tk.Tk()
root.title("Simple Calculator")

equation = tk.StringVar()
entry = tk.Entry(root, textvariable=equation, font=('Arial', 18), bd=8, relief="sunken", justify="right")
entry.grid(columnspan=4, ipadx=8, ipady=8)

# Buttons layout
buttons = [
    ('7',1,0), ('8',1,1), ('9',1,2), ('+',1,3),
    ('4',2,0), ('5',2,1), ('6',2,2), ('-',2,3),
    ('1',3,0), ('2',3,1), ('3',3,2), ('*',3,3),
    ('0',4,0), ('C',4,1), ('=',4,2), ('/',4,3),
]

for (text, row, col) in buttons:
    if text == "=":
        btn = tk.Button(root, text=text, height=2, width=6, command=equalpress)
    elif text == "C":
        btn = tk.Button(root, text=text, height=2, width=6, command=clear)
    elif text in ['+', '-', '*', '/']:
        btn = tk.Button(root, text=text, height=2, width=6, command=lambda t=text: press(" " + t + " "))
    else:
        btn = tk.Button(root, text=text, height=2, width=6, command=lambda t=text: press(t))
    btn.grid(row=row, column=col, padx=5, pady=5)

root.mainloop()
```
